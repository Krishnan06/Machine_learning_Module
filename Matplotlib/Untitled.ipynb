```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Day=[1,2,3,4,5,6,7]
Temp=[50,51,52,48,47,49,46]
```

```python
plt.plot(Day,Temp,color='green',linestyle='dashdot',marker='o',linewidth=2)
```

```python
plt.title("Wether info",fontsize=14)
plt.xlabel('Day',fontsize=12)
plt.ylabel('Temperature')

plt.plot(Day,Temp,color='green',linestyle='dashdot',marker='o',linewidth=2,alpha=0.2)
```

```python
days=[1,2,3,4,5,6,7] 
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37] 
avg_t=[45,48,48,46,40,42,41] 
```

```python
plt.title("Weather info",fontsize=14)
plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc="best",shadow=True)
plt.grid()
```

```python
# Bar chart


company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Top Companies")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Top Companies")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```
