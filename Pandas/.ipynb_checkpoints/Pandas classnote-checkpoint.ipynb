```python
df_customer =pd.DataFrame({ 
    'id':[1,2,3,4], 'name':['Tom', 'Jenny', 'James', 'Dan']
})
df_info = pd.DataFrame({ 
    'id': [2, 3, 4, 5], 
    'age': [31, 20, 40, 70], 
    'sex': ['F', 'M', 'M', 'F']
}) 
```

```python
pd.merge(df_customer,df_info)
```

```python
df_customer.merge(df_info)
```

```python
pd.merge(df_customer,df_info,on='id')
```
